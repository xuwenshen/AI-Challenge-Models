```python
import numpy as np
from tqdm import *
from collections import defaultdict
import jieba

zh_data = open('/data/xuwenshen/ai_challenge/data/train/train/train.zh', encoding='utf-8').readlines()
zh_tokens = []

zhl = defaultdict(lambda : 0)

for i in tqdm(range(len(zh_data))):
    zh_data[i]=zh_data[i].rstrip('\n')
    seg_list = ' '.join(jieba.cut(zh_data[i])).split()
    zhl[len(seg_list)] += 1
    zh_tokens.append(seg_list)
```

```python
# 将en和zh都dump到同一个h5py文件，valid.h5py
import json
import h5py
from collections import defaultdict
from tqdm import *
import random

_en = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/en_data.json'))
en_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/en_voc.json'))

_zh = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/zh_data.json'))
zh_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/zh_voc.json'))

print (len(_en))
print (len(_zh))

data = [0 for i in range(len(_en))]
for i in range(len(data)):
    data[i] = [_en[i], _zh[i]]
random.shuffle(data)

f = h5py.File("/data/xuwenshen/ai_challenge/data/train/train_words/train0.h5py", "a")


en_set = f.create_dataset("en", (len(data), 60), dtype='i8')
en_length_set = f.create_dataset('en_len', (len(data), ), dtype='i8')


label_set = f.create_dataset("label", (len(data), 51), dtype='i8')
zh_length_set = f.create_dataset('zh_len', (len(data), ), dtype='i8')
gtruth_set = f.create_dataset('ground_truth', (len(data), 51), dtype='i8')

zh_str_set = f.create_dataset('zh_str', (len(data), ), dtype='S400')
en_str_set = f.create_dataset('en_str', (len(data), ), dtype='S400')


en = []
en_len = []
en_str = []

labels = []
zh_len = []
gtruths = []
zh_str = []

for i in tqdm(range(len(data))):
    
    #en data
    tmp = [en_voc['pad#'] for j in range(60)]
    if len(data[i][0]) == 0 or len(data[i][0]) > 60:
        print ('en ', len(data[i][0]))
        continue
    if len(data[i][1]) == 0 or len(data[i][1]) > 50:
        print ('zh ', len(data[i][1]))
        continue
    
    for j in range(len(data[i][0])):
        if data[i][0][j].lower() not in en_voc:
            tmp[j] = en_voc['nuk#']
        else:
            tmp[j] = en_voc[data[i][0][j].lower()]
    en.append(tmp)
    en_len.append(len(data[i][0]))
    en_str.append(' '.join(data[i][0][:min(400, len(data[i][0]))]).encode('utf8') )
    
    # zh data
    tlabel = [zh_voc['pad#'] for j in range(51)]
    tgtruth = [zh_voc['pad#'] for j in range(51)]
    for j in range(len(data[i][1])):
        if data[i][1][j] not in zh_voc:
            tlabel[j] = zh_voc['nuk#']
            tgtruth[j+1] = zh_voc['nuk#']
        else:
            tlabel[j] = zh_voc[data[i][1][j]]
            tgtruth[j+1] = zh_voc[data[i][1][j]]
            
    tlabel[len(data[i][1])] = zh_voc['eos#']
    tgtruth[0] = zh_voc['go#']
    
    
    labels.append(tlabel)
    gtruths.append(tgtruth)
    zh_len.append(len(data[i][1])+1)    
    zh_str.append(''.join(data[i][1][:min(200, len(data[i][1]))]).encode('gb2312', 'ignore'))
    
    if len(en) == 256:
        start = i+1-256
        end = i+1
        # en data
        en_set[start : end] = en
        en_length_set[start : end] = en_len
        en_str_set[start : end] =  en_str
        en = []
        en_len = []
        en_str = []
        
        #zh data
        label_set[start : end] = labels
        zh_length_set[start : end] = zh_len
        gtruth_set[start : end] = gtruths      
        zh_str_set[start:end] = zh_str
        labels = []
        zh_len = []
        gtruths = []
        zh_str = []

        
end = len(data)
start = end-len(en)
print (start, end)
en_set[start : end] = en
en_length_set[start : end] = en_len
en_str_set[start : end] =  en_str

label_set[start : end] = labels
zh_length_set[start : end] = zh_len
gtruth_set[start : end] = gtruths
zh_str_set[start : end] =  zh_str
f.close()
```

```python
# 解析valid en.sgm 和 test en.sgm
import json
import xml
import re
import nltk

fin = open('/data/xuwenshen/ai_challenge/data/test/test.sgm').readlines()

start = re.compile(r'\d+\"> ')
end = re.compile(r' </seg>')

en = []
for i in range(len(fin)):
    
    lst = start.split(fin[i])
    if len(lst) != 2:
        print (fin[i])
        continue
    
    lst = end.split(lst[-1])
    
    if len(lst) != 2:
        print (fin[i])
        continue
    
    en.append(lst[0])

    
    
en_tokens = []
lens = []
for i in range(len(en)):
    tokens = nltk.word_tokenize(en[i])
    en_tokens.append(tokens)
    lens.append(len(tokens))
    
fout = open('/data/xuwenshen/ai_challenge/data/test/test_words/en.json', 'w')

json.dump(en_tokens, fout)

fout.close()
print (max(lens))
print (min(lens))
```

```python
# 解析valid zh.sgm
import json
import xml
import re
import nltk
import jieba


fin = open('/data/xuwenshen/ai_challenge/data/valid/valid/valid.en-zh.zh.sgm').readlines()

start = re.compile(r'\d+\"> ')
end = re.compile(r' </seg>')

zh = []
for i in range(len(fin)):
    
    lst = start.split(fin[i])
    if len(lst) != 2:
        print (fin[i])
        continue
    
    lst = end.split(lst[-1])
    
    if len(lst) != 2:
        print (fin[i])
        continue
    
    zh.append(lst[0])

zh_tokens = []
lens = []

for i in range(len(zh)):
    
    zh[i]=zh[i].rstrip('\n')
    seg_list = ' '.join(jieba.cut(zh[i])).split()
    zh_tokens.append(seg_list)
    lens.append(len(seg_list))
    
fout = open('/data/xuwenshen/ai_challenge/data/valid/valid_words/zh.json', 'w')

json.dump(zh_tokens, fout)

fout.close()
print (max(lens))
print (min(lens))
```

```python
# 将en和zh都dump到同一个h5py文件，valid.h5py
import json
import h5py
from collections import defaultdict
from tqdm import *

_en = json.load(open('/data/xuwenshen/ai_challenge/data/valid/valid_words/en.json'))
en_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/en_voc.json'))

_zh = json.load(open('/data/xuwenshen/ai_challenge/data/valid/valid_words/zh.json'))
zh_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train_words/zh_voc.json'))

print (len(_en))
print (len(_zh))

f = h5py.File("/data/xuwenshen/ai_challenge/data/valid/valid_words/valid.h5py", "a")


en_set = f.create_dataset("en", (len(_en), 60), dtype='i8')
en_length_set = f.create_dataset('en_len', (len(_en), ), dtype='i8')


label_set = f.create_dataset("label", (len(_zh), 51), dtype='i8')
zh_length_set = f.create_dataset('zh_len', (len(_zh), ), dtype='i8')
gtruth_set = f.create_dataset('ground_truth', (len(_zh), 51), dtype='i8')

zh_str_set = f.create_dataset('zh_str', (len(_zh), ), dtype='S400')
en_str_set = f.create_dataset('en_str', (len(_en), ), dtype='S400')


en = []
en_len = []
en_str = []

labels = []
zh_len = []
gtruths = []
zh_str = []

for i in tqdm(range(len(_en))):
    
    #en data
    tmp = [en_voc['pad#'] for j in range(60)]
    if len(_en[i]) == 0 or len(_en[i]) > 60:
        print ('en ', len(_en[i]))
        continue
    if len(_zh[i]) == 0 or len(_zh[i]) > 50:
        print ('zh ', len(_zh[i]))
        continue
    
    for j in range(len(_en[i])):
        if _en[i][j].lower() not in en_voc:
            tmp[j] = en_voc['nuk#']
        else:
            tmp[j] = en_voc[_en[i][j].lower()]
    en.append(tmp)
    en_len.append(len(_en[i]))
    en_str.append(' '.join(_en[i][:min(400, len(_en[i]))]).encode('utf8') )
    
    # zh data
    tlabel = [zh_voc['pad#'] for j in range(51)]
    tgtruth = [zh_voc['pad#'] for j in range(51)]
    for j in range(len(_zh[i])):
        if _zh[i][j] not in zh_voc:
            tlabel[j] = zh_voc['nuk#']
            tgtruth[j+1] = zh_voc['nuk#']
        else:
            tlabel[j] = zh_voc[_zh[i][j]]
            tgtruth[j+1] = zh_voc[_zh[i][j]]
            
    tlabel[len(_zh[i])] = zh_voc['eos#']
    tgtruth[0] = zh_voc['go#']
    
    
    labels.append(tlabel)
    gtruths.append(tgtruth)
    zh_len.append(len(_zh[i])+1)    
    zh_str.append(''.join(_zh[i][:min(200, len(_zh[i]))]).encode('gb2312', 'ignore'))
    
    if len(en) == 256:
        start = i+1-256
        end = i+1
        # en data
        en_set[start : end] = en
        en_length_set[start : end] = en_len
        en_str_set[start : end] =  en_str
        en = []
        en_len = []
        en_str = []
        
        #zh data
        label_set[start : end] = labels
        zh_length_set[start : end] = zh_len
        gtruth_set[start : end] = gtruths      
        zh_str_set[start:end] = zh_str
        labels = []
        zh_len = []
        gtruths = []
        zh_str = []

        
end = len(_en)
start = end-len(en)
print (start, end)
en_set[start : end] = en
en_length_set[start : end] = en_len
en_str_set[start : end] =  en_str

label_set[start : end] = labels
zh_length_set[start : end] = zh_len
gtruth_set[start : end] = gtruths
zh_str_set[start : end] =  zh_str
f.close()
```
