```python
# 将en和zh都dump到同一个h5py文件，valid.h5py, train.h5py
import json
import h5py
from collections import defaultdict
from tqdm import *
import random

_en = json.load(open('/data/xuwenshen/ai_challenge/data/train/train/en_data-50-60.json'))
en_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train/en_voc.json'))
print ('en finished...')

_zh = json.load(open('/data/xuwenshen/ai_challenge/data/train/train/zh_data-50-60.json'))
zh_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train/zh_voc.json'))
print ('zh finished...')

print (len(_en))
print (len(_zh))

data = []

for i in range(len(_en)):
    data.append([_en[i], _zh[i]])

# shuffle isf train data
random.shuffle(data)

f = h5py.File("/data/xuwenshen/ai_challenge/data/train/train/ibm_train-50-60.h5py1", "a")


en_set = f.create_dataset("en", (len(data), 50), dtype='i8')
en_length_set = f.create_dataset('en_len', (len(data), ), dtype='i8')


label_set = f.create_dataset("label", (len(data), 61), dtype='i8')
zh_length_set = f.create_dataset('zh_len', (len(data), ), dtype='i8')
gtruth_set = f.create_dataset('ground_truth', (len(data), 62), dtype='i8')

zh_str_set = f.create_dataset('zh_str', (len(data), ), dtype='S400')
en_str_set = f.create_dataset('en_str', (len(data), ), dtype='S400')


en = []
en_len = []
en_str = []

labels = []
zh_len = []
gtruths = []
zh_str = []

for i in tqdm(range(len(data))):
    
    #en data
    tmp = [en_voc['pad#'] for j in range(50)]
    if len(data[i][0]) == 0 or len(data[i][0]) > 50:
        print ('en ', len(data[i][0]))
        data[i][0] = data[i][0][:50]
        continue
    if len(data[i][1]) == 0 or len(data[i][1]) > 60:
        print ('zh ', len(data[i][1]))
        data[i][1] = data[i][1][:60]
        continue
    
    for j in range(len(data[i][0])):
        if data[i][0][j].lower() not in en_voc:
            tmp[j] = en_voc['nuk#']
        else:
            tmp[j] = en_voc[data[i][0][j].lower()]
    en.append(tmp)
    en_len.append(len(data[i][0]))
    en_str.append(' '.join(data[i][0][:min(400, len(data[i][0]))]).encode('utf8'))
    
    # zh data
    tlabel = [zh_voc['pad#'] for j in range(61)]
    tgtruth = [zh_voc['pad#'] for j in range(62)]
    for j in range(len(data[i][1])):
        if data[i][1][j] not in zh_voc:
            tlabel[j] = zh_voc['nuk#']
            tgtruth[j+1] = zh_voc['nuk#']
        else:
            tlabel[j] = zh_voc[data[i][1][j]]
            tgtruth[j+1] = zh_voc[data[i][1][j]]
            
    tlabel[len(data[i][1])] = zh_voc['eos#']
    tgtruth[0] = zh_voc['go#']
    tgtruth[len(data[i][1])] = zh_voc['eos#']
    
    
    labels.append(tlabel)
    gtruths.append(tgtruth)
    zh_len.append(len(data[i][1])+1)    
    zh_str.append(''.join(data[i][1][:min(200, len(data[i][1]))]).encode('gb2312', 'ignore'))
    
    if len(en) == 256:
        start = i+1-256
        end = i+1
        # en data
        en_set[start : end] = en
        en_length_set[start : end] = en_len
        en_str_set[start : end] =  en_str
        en = []
        en_len = []
        en_str = []
        
        #zh data
        label_set[start : end] = labels
        zh_length_set[start : end] = zh_len
        gtruth_set[start : end] = gtruths      
        zh_str_set[start:end] = zh_str
        labels = []
        zh_len = []
        gtruths = []
        zh_str = []

        
end = len(data)
start = end-len(en)
print (start, end)
en_set[start : end] = en
en_length_set[start : end] = en_len
en_str_set[start : end] =  en_str

label_set[start : end] = labels
zh_length_set[start : end] = zh_len
gtruth_set[start : end] = gtruths
zh_str_set[start : end] =  zh_str
f.close()
```

```python
import h5py
import json
from collections import defaultdict

zh_voc = json.load(open('/data/xuwenshen/ai_challenge/data/train/train/zh_voc.json'))

h5 = h5py.File('/data/xuwenshen/ai_challenge/data/train/train/ibm_train-50-60.h5py', 'r')

rvoc = defaultdict()
for (k,v) in zh_voc.items():
    rvoc[v] = k
rvoc = dict(rvoc)

f1 = h5['ground_truth']
f2 = h5['label']
f3 = h5['zh_str']
for i in range(100):
    text1 = ''
    text2 = ''
    for j in range(62):
        if f1[i][j] == zh_voc['pad#']: 
            break
        text1 += rvoc[f1[i][j]]
    for j in range(61):
        if f2[i][j] == zh_voc['pad#']:
            break
        text2 += rvoc[f2[i][j]]
    
    
    print (i, 'ground_truth ', text1)
    print (i, 'label ', text2)
    print (i, 'zh_str ', f3[i].decode('gb2312'))
    
```
